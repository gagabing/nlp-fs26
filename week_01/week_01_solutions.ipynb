## Byte-pair encoding

Implement Byte-paid encoding and reproduce the example in chapter 2.4.1 of [Speech and Language Processing](https://web.stanford.edu/~jurafsky/slp3/):

```python
corpus = ["lowest_", "low_", "newer_", "wider_", "new_"]
corpus = list(map(lambda word: [letter for letter in word], corpus))
corpus
```

```python
def create_v(corpus):
    V = []
    for word in corpus:
        for letter in word:
            V.append(letter)

    V = set(V)
    return V
```

```python
V = create_v(corpus)
V
```

```python
from nltk import bigrams, FreqDist

def count_adjacent(corpus):
    adjacent = []
    for word in corpus:
        adjacent.extend(list(bigrams(word)))

    freq = FreqDist(adjacent)
    freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)
    return freq

def get_top(adjacent):
    return ''.join(adjacent[0][0])
```

```python
adjacent = count_adjacent(corpus)
adjacent
```

```python
top = get_top(adjacent)
top
```

```python
def replace_top_in_seq(top, seq):
    new_word = []

    if len(seq) == 1:
        return seq
    
    pairs = list(zip(seq, seq[1:]))
    
    while len(pairs) > 0:
        left, right = pairs.pop(0)
        pair = ''.join([left, right])

        if pair == top:
            new_word.append(pair)
            if len(pairs) > 1:
                _ = pairs.pop(0)
            elif len(pairs) == 1:
                _, right = pairs.pop(0)
                new_word.append(right)
        elif len(pairs) == 0:
            new_word.append(left)
            new_word.append(right)
        else:
            new_word.append(left)

    return new_word


def replace_top_in_corpus(top, corpus):
    new_corpus = []

    for seq in corpus:
        new_seq = replace_top_in_seq(top, seq)
        new_corpus.append(new_seq)
    
    return new_corpus


def byte_pair_encoding(corpus, k, verbose = True):

    C = list(map(lambda word: [letter for letter in word], corpus))
    V = create_v(corpus)

    for iter_k in range(k):
        adjacent = count_adjacent(C)
        top = get_top(adjacent)
        V.add(top)
        C = replace_top_in_corpus(top, C)
        if verbose: print(f"{iter_k+1}: \t {top=} \t {C=}")
    
    return C, V
    
```

```python
corpus = ["lowest_", "lowest_", "low_", "low_", "low_", "low_", "low_", "newer_", "newer_", "newer_", "newer_", "newer_", "newer_", "wider_", "wider_", "wider_", "new_", "new_"]
C, V = byte_pair_encoding(corpus, 16)
```

# Tokenizer

Go through the HuggingFace tutorial on [tokenizers](https://huggingface.co/learn/nlp-course/chapter2/4) and answer:

- why do we need tokenizers?
- what is the difference between a character-based, word-based and a subword-based tokenizer? What are the advantages and disadvantages of each?

Answer: Why tokenizers?
- mapping from raw text to machine readable format
- tokenization is the first step in the NLP pipeline and we need numbers not symbols for ML models
- breaks down text into smaller units (called tokens) which preserves the meaning of the text - ideally we efficently and compactly represent the text
- each token corresponds to a numerical value (IDs)
- help manage the total size of the vocabulary
- help dealing with OVV

Answer: Difference between character-based, word-based and subword-based tokenizer
- character-based: tokenizes the text into characters
    - advantage: can handle any language, small vocabulary size, no OVV, easy to implement
    - disadvantage: one word is made of many tokens, given a token does not say too much of the token before/after (little semantic meaning), slow to train, NNs have to have a huge context window (later)
- word-based: tokenizes the text into words (using spaces or punctuation)
    - advantage: more semantic meaning, faster to train, NNs can have smaller context window, easy to implement
    - disadvantage: OVV, different forms of the same word are treated as different tokens (e.g. dog/dogs), language dependent, large vocabulary size
- subword-based: tokenizes the text into subwords (e.g. Byte-Pair Encoding). Most widely used tokenizers
    - advantage: can handle OVV, more compact vocabulary size than word-based, can handle different forms of the same word, works well for languages with complex morphology (e.g. German or Turkish)
    - disadvantage: more complex to implement, slower to train, might loose semantic meaning, rare words might be splitted into many tokens, computationally more expensive
